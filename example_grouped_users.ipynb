```python
import pandas as pd
import numpy as np
from pathlib import Path
import map_route as mp
import itertools
import folium
from folium import Marker
```

```python
data_points = [
    ["UNAL", 6.2623506, -75.5778091],
    ["UPB", 6.2536481, -75.5815857],
    ["IUE", 6.1621011, -75.5944852],
]
points = pd.DataFrame(data_points, columns=["id", "Latitud", "Longitud"])
```

Se crea una función para generalizar generar datos aleatorios sobre agrupamientos

```python
def gen_users(to_lat, to_lon, n_users):
    table = []
    for _ in range(n_users):
        ran_lat, ran_lon = np.random.normal(size=2) * .001
        table.append([to_lat + ran_lat, to_lon + ran_lon, to_lat, to_lon])
    return table
```

Se crean datos ficticios para mostrar 2 personas agrupadas en la estación 0 y 3 en 1

```python
stations_loc = [0, 1]
n_persons_stations = [2, 3]
acc_table = []
for stat, n_persons in zip(stations_loc, n_persons_stations):   
    to_lat, to_lon = points.iloc[stat, 1:]
    acc_table.append(gen_users(to_lat, to_lon, n_persons))

table_grouped = itertools.chain.from_iterable(acc_table)
grouped_users = pd.DataFrame(table_grouped, columns=["from_lat", "from_lon", "to_lat", "to_lon"])
grouped_users["id"] = ["id" +  str(x) for x in range(len(grouped_users))]
grouped_users.head(2)
```

```python
test_route = mp.get_route(points)
```

```python
m = mp.map_route(test_route)
mp.add_grouped_users(m, grouped_users)
m
```
